```python
### NOVO CONTATO 
class Contatos:
    def __init__(self, nome, telefone, email):
        self.nome = nome
        self.telefone = telefone
        self.email = email
        print('Novo contato {} criado com sucesso!'.format(self.nome))
        adicionar_arq(self.nome, self.telefone, self.email)
        
#tratamento de entradas
def tratamento_nome(ent_n):
    ent_n = ent_n.title()
    ent_n = ent_n.strip()
    return ent_n

def validacao_tel(ent_t):
    if 10 <= len(ent_t) <= 11 and ent_t.isdigit() == True:
        pass
    else:
        ent_t = 0
    return ent_t

def validacao_email(ent_e):
    if ent_e.find('@') >= 1 and (ent_e.endswith('.com') >= 1 or ent_e.endswith('.com.br') >= 1 or ent_e.endswith('.org') >= 1 or ent_e.endswith('.net') >= 1):
        pass
    else:
        ent_e = 0
    return ent_e
        
    
def entrada_novo_contato():
    ent_n = input('Digite o nome: ')
    ent_n = tratamento_nome(ent_n)

    while True:
        ent_t = input('Digite o telefone [apenas números: 2 para DDD + 8 para fixo ou 9 para celular]: ')
        if validacao_tel(ent_t) == 0:
            print('Telefone em formato inválido.')
        else:
            break

    while True:
        ent_e = input('Digite o e-mail: ')
        if validacao_email(ent_e) == 0:
            print('E-mail inválido.')
        else:
            break

    novo = Contatos(ent_n, ent_t, ent_e)
    
### ARQUIVOS

l_arq = []

def carregar_arq():
    if x == 1:
        arq = open(r'C:\Users\Denise_\Documents\Pi_DS\Módulo 1\Arqs\contatos.csv', 'r')
    else:
        arq = open(r'C:\Users\Denise_\Documents\Pi_DS\Módulo 1\Arqs\contatos_org.csv', 'r')
    for i in arq:
        l_arq.append(i.split(';'))
    arq.close()
    return l_arq

def ordenar_arq(novo):
    carregar_arq()
    if novo == 0:
        pass
    else:
        l_arq.append(novo)
    cont = 0
    for i in l_arq:
        for j in i:
            jj = j.strip()
            if jj.isdigit() == True:
                tel = jj
            elif jj.find('@') >= 1:
                ema = jj
            elif len(jj) > 2:
                nom = jj
            else:
                print('Ops! Problema encontrado na linha {} do arquivo.'.format(cont + 1))
        l_arq[cont][0] = ''.join([nom, ';'])
        l_arq[cont][1] = ''.join([tel, ';'])
        l_arq[cont][2] = ''.join([ema, '\n'])
        cont += 1     
    
def gravar_arq(l_arq):
    arq = open(r'C:\Users\Denise_\Documents\Pi_DS\Módulo 1\Arqs\contatos_org.csv', 'w')
    for i in l_arq:
        for j in i:
            arq.write(j)
    arq.close()
    l_arq.clear()
    print('Salvo com sucesso!')
    
def adicionar_arq(nome, telefone, email):
    novo = []
    novo.append(nome)
    novo.append(telefone)
    novo.append(email)
    ordenar_arq(novo)
    print('Adicionado com sucesso!')
    gravar_arq(l_arq)

def consultar_contato():
    n = input('Digite o nome que quer consultar: ')
    n = ''.join([n, ';'])
    c = 0
    c1 = 999
    ordenar_arq(0)
    for i in l_arq:
        for j in i:
            if n == j:
                c1 = c
        c += 1
    if c1 == 999:
        print('Não encontrado. Tente novamente.')
    else:
        print('Nome: {}  Telefone: {}  E-mail: {}'.format(l_arq[c1][0], l_arq[c1][1], l_arq[c1][2]))
    l_arq.clear()
    
def remover_contato():
    n = input('Digite o nome que quer deletar: ')
    n = ''.join([n, ';'])
    c = 0
    c1 = 999
    ordenar_arq(0)
    for i in l_arq:
        for j in i:
            if n == j:
                c1 = c
        c += 1
    if c1 == 999:
        print('Não encontrado. Tente novamente.')
    else:
        del l_arq[c1]
        print('Contato deletado!')
        gravar_arq(l_arq)
    l_arq.clear()



### TELA INICIAL

print('CADASTRO DE FORNECEDORES')

while True:
    try:
        x = int(input('''Digite a operação:
              1 - Carregar e organizar arquivo
              2 - Novo contato
              3 - Remover contato
              4 - Consultar contato
              0 - Sair\n'''))
        
        if x == 1:
            ordenar_arq(0)
            gravar_arq(l_arq)
        elif x == 2:
            entrada_novo_contato()
        elif x == 3:
            remover_contato()
        elif x == 4:
            consultar_contato()
        elif x == 0:
            print('Até logo! :)')
            break
    except Exception as erro:
        print('Ops! Comando inválido!')
        continue
```
